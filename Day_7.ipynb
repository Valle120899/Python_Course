
## Hangman

```python
import random

Fail_1 = """
|
|
|
|
"""

Fail_2 = """
|------
|
|
|
"""

Fail_3 = """
|------
|     |
|
|
"""

Fail_4 = """
|------
|     |
|     O
|
|
|
|
"""

Fail_5 = """
|------
|     |
|     O
|     |
|
|
|
"""

Fail_6 = """
|------
|     |
|     O
|    -|
|
|
|
"""

Fail_7 = """
|------
|     |
|     O
|    -|-
|
|
|
"""

Fail_8 = """
|------
|     |
|     O
|    -|-
|    /
|
|
"""

Fail_9 = """
|------
|     |
|     O
|    -|-
|    / \
|
|
"""
Fails = [Fail_1,Fail_2,Fail_3,Fail_4,Fail_5,Fail_6,Fail_7,Fail_8,Fail_9]

Secret_words = ['ardvark', 'baboon', 'camel']

Value = random.randint(0, len(Secret_words)-1)

Word = Secret_words[Value]
Game_List = []
for i in range(len(Word)):
    Game_List.append('_')

Life = -1
Aux= 0

#Game
print('Welcome to the hangman game!')
while Life != 8: #8 is the max life per game
    print(Game_List)
    Opt = input('Try to guess a unique letter!\n')
    if Opt in Word:
        for i in range(len(Word)):
            if Word[i] == Opt:
                Game_List[i] = Opt
                Aux+=1
    else:
        print('Upsi')
        Life+=1
        print(f"{Fails[Life]}")
    if Aux == len(Game_List):
        break
                
if Life == 8:
    print('You lost!')
else:
    print('You won!')
```
